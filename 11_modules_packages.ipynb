# Modules

While using the interpreter you may notice that the state (functions and variables) is lost between sessions.  While this may suffice for short programs you will likely have issues when attempting larger projects or if you wish to rerun segments of code.  You can save your definitions in text files to be used later in a script or an interactive instance of the interpreter.  These files are called *modules*.  A module is a file containing Python definitions and statements.  Definitions from a module can be *imported* into other modules or into the *main* module.  The main module is the collection of variables that you have access to in a script executed at the top level.  

The name of a module is the same as the file name (with suffix `.py`).  

As an example the following code can be found in fibo.py

```
# Fibonacci numbers module

def fib(n):    # write Fibonacci sequence up to n
    a, b = 0, 1
    while a < n:
        print(a, end=' ')
        a, b = b, a+b
    print()

def fib2(n):   # return Fibonacci sequence up to n
    result = []
    a, b = 0, 1
    while a < n:
        result.append(a)
        a, b = b, a+b
    return result
```

## `import`

To use the functions from the above module we import them as follows:

````python
import fibo
````

In the above example we import the fibo module.  By default the import statement will enter only the name of the module into the global namespace (not all of its contained names).  Thus to call a function from the imported module you must use the module name followed by a `.` and then the function name, as in the following example.

````python
fibo.fib(1000)
````

It is recommended that all imports to appear at the beginning of the file or script.

## `import` Variants

Instead of importing all of the definitions found in a module you can import selected definitions.

````python
from fibo import fib
fib(200)
````

This example does two things differently from the previous `import fibo`. 
- Only the function `fib` is imported from fibo.
- The function `fib` is entered directly into the global namespace instead of the module name `fibo`.  Thus we can call `fib` directly without the module name as we did previously, `fibo.fib`.  

This variant of the import statement can be used to import all definitions using `*` as follows:

````python
from fibo import *
fib(400)
````

Using `*` imports all names except those beginning with `_`.  Typically Python programmers do not use this facility (*) as it introduces an unknown set of names into the interpreter, which may hide some things already defined. 

## `import` Aliases

Aliases can be defined for imports using `as`.  For example:

````python
import fibo as fib
fib.fib(400)
````

The above example imports the fibo module but uses the name fib in the global namespace in place of fibo.  This can be used to save keystrokes for modules with long names or to improve the readability of your code.  Aliases can also be applied to imported definitions as follows:

````python
from fibo import fib as fibonacci
fibonacci(400)
````

## The `dir()` Function

The built-in `dir()` function is used to find out what definitions are available in a module.

````python
dir(fibo)
````

Without any arguments the `dir()` function returns the names that are currently defined.

````python
dir()
````

# [Packages](https://docs.python.org/3/reference/import.html#packages)

Packages are Pythons way of organizing module using directories and  "dotted module names".  For example, `A.B` denotes submodule `B` in package `A`. 

Let us consider the following package for handling sounds files:

```
sound/                          Top-level package
      __init__.py               Initialize the sound package
      formats/                  Subpackage for file format conversions
              __init__.py
              wavread.py
              wavwrite.py
              aiffread.py
              aiffwrite.py
              auread.py
              auwrite.py
              ...
      effects/                  Subpackage for sound effects
              __init__.py
              echo.py
              surround.py
              reverse.py
              ...
      filters/                  Subpackage for filters
              __init__.py
              equalizer.py
              vocoder.py
              karaoke.py
              ...
```

The top directory, sound, contains several subdirectories that organize the functionality of the package: reading and writing various formats (`sound/formats/`), processing sounds data to apply effects (`sound/effects/`) or filtering (`sound/filters/`), etc.  Within each directory a special file, `__init__.py`, exists that tells Python the directory is to be considered part of a package.  This prevents directories with common names from unintentionally hiding valid modules that appear later in the search path.

## `import`

```
import sound.effects.echo
```
The above imports the echo module from the sound.effects package.  The full name must be used in order to reference contained functions, for example:
```
sound.effects.echo.echofilter(input, output, delay=0.7, atten=4)
```
As noted at the end of the [`import` Variants](#import-Variants) section above all items that appear in the import must be a package name, except for the last which can be either a module or a package.  In the above example, both `sound` and `effects` are packages (or subpackage) and `echo` is the module stored in the file `echo.py`.

Submodules can also be imported using the import variant: 
```
from sound.effects import echo
```
Aliases can be used for packages just as they were for modules and their definitions in [`import` Aliases](#import-Aliases).

### Importing * From a Package

Using this is considered bad practice in production code.

Unlike with modules, when `from package import *` is encountered the entirety of the package is not imported.  If this were the case it would potentially take a long time to import every aspect of the package and it may have some unwated side-effects that should only happen when the sub-module is explicitly imported.  Instead Python looks for `__all__` to be defined in `__init__.py`.  The variable `__all__`, if defined, is a list of subpackages or submodules to be imported, for example:
```
__all__ = ["echo", "surround", "reverse"]
```
This would allow `from sound.effects import *` to import the three named modules.  If `__all__` is not defined, the package name itself is entered into the global namespace and any names explicitly loaded in its `__init__.py`.



## `__init__.py`

`__init__.py` may be empty, but may contain any initialization code required by the package.  This includes any code that a regular module can contain, plus some special variables like `__all__`.  For more information read the documentation on [regular packages](https://docs.python.org/3/reference/import.html#regular-packages) and the [import system](https://docs.python.org/3/reference/import.html#the-import-system).
