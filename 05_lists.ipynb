# Python Lists

ds_list.png

A [python list](https://docs.python.org/3.8/library/stdtypes.html#list) is a compound data type, which is used to group other values.  Lists are written as a sequence of comma-separated values (items) between square brackets.  The items of a list may be different types (heterogeneous data), but typically they are of the same type.  Lists are useful when the order of elements matter or when elements can be addressed by a numeral (index).



```python
squares = [1, 4, 9, 16, 25]
squares
```

```python
days_of_the_week = ['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
days_of_the_week
```

```python
mixed_bag = ['Space', 42, squares]
mixed_bag
```

## Indexing

Lists may be indexed (subscripted) just like all other built-in [sequence](https://docs.python.org/3.8/glossary.html#term-sequence) types.  Python uses zero-based indices, meaning that the first element has index 0, the second has index 1, and the last has index n-1.

To index a list we use the square bracket notation, which immediately follows the list variable and encloses a zero-based index.


```python
squares
```

```python
squares[0]
```

```python
squares[1]
```

```python
squares[4]
```

Accessing a list using an index beyond the length of the list results in an error.

```python
squares[5]
```

```python
len(squares)
```

Using negative indices results in a right-to-left indexing.

```python
squares[-1]
```

```python
squares[-5]
```

```python
squares[-6]
```

## Slicing
Lists may be sliced just like all other built-in [sequence](https://docs.python.org/3.8/glossary.html#term-sequence) types.  

All slice operations return a new list containing a [shallow copy](https://docs.python.org/3.8/library/copy.html#shallow-vs-deep-copy) of the requested elements.

```python
squares
```

```python
squares[1:3]
```

```python
squares[1:1]
```

```python
squares[1:2]
```

Slices may contain negative indices

```python
squares[1:-1]
```

```python
squares[-3:4]
```

You may omit one of the indices to inlcude the beginning or end of the list.

```python
squares[:-3]
```

```python
squares[1:]
```

Slicing also supports a step size

```python
squares[1:5:2]
```

In the indexing section above we saw that accessing an index that is out of range results in an error. Slicing handles out of range indices a bit more gracefully.

```python
squares[4:42]
```

```python
squares[42:]
```

  ## Concatenation
  
  Lists may be concatenated using the `+` operator.

```python
squares = [1, 4, 9, 16, 25]
squares
```

```python
squares + [36, 49, 64, 81, 100]
```

```python
squares
```

```python
squares = squares + [36]
squares
```

```python
squares += [49, 64]
squares
```

The append and extend list methods can also be used.

```python
squares.append(81) # append a single element to the list
squares
```

```python
squares.extend([100, 121, 144]) # extend the list with all elements in the passed list
squares
```

Lists can be repeated using the `*` operator.

```python
[4, 2] * 5
```

## Mutability

Unlike strings, lists are [mutable](https://docs.python.org/3.8/glossary.html#term-mutable), meaning that their contents can be changed.  The values of a list can be modified using their index, slicing and even through methods like [`append()`](https://docs.python.org/3/tutorial/datastructures.html#more-on-lists) (functions will be covered later).

```python
cubes = [1, 8, 27, 65, 125] # 65 is incorrect here
cubes
```

```python
cubes[3] = 4**3
cubes
```

```python
letters = ['a', 'b', 'C', 'D', 'f']
letters
```

```python
letters[2:4] = ['c', 'd', 'e'] # replace some elements and insert a new one
letters
```

```python
letters[:3] = [] # remove some elements
letters
```
