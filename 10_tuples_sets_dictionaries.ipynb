# Tuples

ds_tuple.png

A tuple is a standard sequence data type ([Sequence Types](https://docs.python.org/3.8/library/stdtypes.html#typesseq)), which consists of a number of comma separated values.  

## Creation

To create a tuple use parentheses containing a comma seperated list of elements.

````python
t = (12345, 54321, 'hello!')
t
````

To create a tuple with zero elements use an empty set of parentheses:

````python
empty = ()
print(type(empty))
len(empty)
````

````python
print(empty)
````

When creating a tuple with only a single element it is insufficient to use parentheses, instead a trailing comma is required. 

````python
non_tuple = ('hello')
print(type(non_tuple))
non_tuple
````

````python
len(non_tuple) # returns the length of the string since this is not a tuple
````

````python
single_tuple = ('hello', )  # <-- note trailing comma
print(type(single_tuple))
single_tuple
````

````python
len(single_tuple)  # returns the number of items in the tuple
````

## Indexing and Slicing

Similar to lists, tuples can be indexed and sliced using the same square bracket syntax.

````python
t[0]
````

````python
t[0:2]
````

## Supported Elements

Tuples may contain any type, including other tuples.

````python
u = (t, (1, 2, 3, 4, 5)) # Tuples may be nested
u
````

## Immutability

Tuples are [immutable](https://docs.python.org/3.8/glossary.html#term-immutable), i.e. it is not possible to assign to the individual items of a tuple:

````python
t = (12345, 54321, 'hello!')
t[0] = 88888  # tuples are immutable
````

However, tuples may contain [mutable](https://docs.python.org/3.8/glossary.html#term-mutable) objects that may be modified:

````python
v = ([1,2,3], [3,2,1]) # tuples can contain mutable objects, like lists
v
````

````python
v[0][1] = 15  # assignment to a list is ok
v
````

## Packing and Unpacking

The statement `t = 12345, 54321, 'hello!'` is an example of *tuple packing*.  The values `12345`, `54321`, and `'hello!'` are packed into the tuple `t`.  The opposite operation, *tuple unpacking* is also possible.

````python
t = 12345, 54321, 'hello!'  # packing
t
````

````python
x,y,z = t  # unpacking
print('x =', x)
print('y =', y)
print('z =', z)
````

These operations are more generally known as *sequence unpacking* and works for any sequence on the right-hand side.  In order to do this, however, you must have the as many variables on the left-hand side of the equals as there are elements in the sequence on the right-hand side.  

# Sets

ds_set.png



A set is an unordered collection with no duplicate elements, and support operations like union, intersection, difference and symmetric difference.  Sets are commonly used for membership testing and eliminating duplicate entries.


## Creation

To create a set in python use either the [set()](https://docs.python.org/3.8/library/stdtypes.html#set) function or curly braces. 

````python
basket = {'apple', 'orange', 'apple', 'pear', 'orange', 'banana'}
print(basket)  # shows that duplicates have been removed
````

````python
basket = set(['apple', 'orange', 'apple', 'pear', 'orange', 'banana'])
print(basket)
````

Note that to create an empty set you must use `set()` and not `{}`, as the latter creates an empty dictionary.

````python
d = {}
print(type(d))
````

````python
s = set()
print(type(s))
s
````

When creating sets consisting of a single element, be aware that the `set()` function expects an iterable, and so the following do not both create a set with a single element.

````python
s = {'asdf'}
s
````

````python
s = set('asdf')
s
````

The `set()` function expects a single argument that is iterable, so the string in the previous example is used to create a set containing four elements.  To create a set containing the string as a single element, one could wrap the string with a list or some other iterable object.

````python
s = set(['asdf'])
s
````

## Indexing and Slicing

Sets are unordered and so they do not support indexing or slicing.

````python
basket[0]
````

## Supported Elements

Sets may only contain types that are [hashable](https://docs.python.org/3.10/glossary.html#term-hashable), meaning objects that never change during their lifetime.  

From the last paragraph of the definition of [hashable](https://docs.python.org/3.10/glossary.html#term-hashable).
```
Most of Pythons immutable built-in objects are hashable; mutable containers (such as lists or dictionaries) are not; immutable containers (such as tuples and frozensets) are only hashable if their elements are hashable. Objects which are instances of user-defined classes are hashable by default. They all compare unequal (except with themselves), and their hash value is derived from their id().
```

Also, check out [`__hash__()`](https://docs.python.org/3/reference/datamodel.html#object.__hash__) for more details.

````python
s = {1,2,3, [1,2,3]}
s
````

````python
s = {1,2,3, (1,2,3)}
s
````

````python
s = {1,2,3, (1,2,[3])}
s
````

## Membership Testing

Use `in` for fast membership testing:

````python
basket
````

````python
'orange' in basket  # fast membership testing
````

````python
'crabgrass' in basket
````

## Operations on Sets

Sets are objects and have a number of methods and operators that can act on them.  More information on available [set functions](https://docs.python.org/3.10/library/stdtypes.html#set) can be found in the Python documentation, including:

- **isdisjoint(other)**: Return True if the set has no elements in common with other. Sets are disjoint if and only if their intersection is the empty set.
- **issubset(other)**: Test whether every element in the set is in other.
- **set <= other**: Test whether every element in the set is in other.
- **set < other**: Test whether the set is a proper subset of other, that is, set <= other and set != other.
- **issuperset(other)**: Test whether every element in other is in the set.
- **set >= other**: Test whether every element in other is in the set.
- **set > other**: Test whether the set is a proper superset of other, that is, set >= other and set != other.
- <b>union(*others)</b>: Return a new set with elements from the set and all others.
- **set | other | ...**: Return a new set with elements from the set and all others.
- <b>intersection(*others)</b>: Return a new set with elements common to the set and all others.
- **set & other & ...**: Return a new set with elements common to the set and all others.
- <b>difference(*others)</b>: Return a new set with elements in the set that are not in the others.
- **set - other - ...**: Return a new set with elements in the set that are not in the others.
- **symmetric_difference(other)**: Return a new set with elements in either the set or other but not both.
- **set ^ other**: Return a new set with elements in either the set or other but not both.
- **copy()**: Return a shallow copy of the set.
- others...

As an example the following creates two sets, `a` and `b`.  Notice that the input strings contain duplicates, and the resulting sets do not.

````python
# Demonstrate set operations on unique letters from two words
a = set('abracadabra')
b = set('alacazam')
a       # unique letters in a
````

````python
b       # unique letters in b
````

Sets support the basic set of set functions, union (`|`), intersection (`&`), difference (`-`) and symmetric difference (`^`). 

````python
a | b  # letters in a or b or both, union
````

````python
a & b  # letters in both a and b, intersection
````

````python
a - b  # letters in a but not in b, difference
````

````python
a ^ b  # letters in a or b but not both, symmetric difference
````

# Set Comprehensions

Comprehensions are supported on sets just like lists.  The only difference is the use of curly braces in place of square brackets.  Refer to the lists_revisited notebook for more details.

````python
a = {x for x in 'abracadabra' if x not in 'abc'}
a
````

# Dictionaries

ds_dictionary.png



A Python *dictionary* is a [Mapping Type](https://docs.python.org/3.10/library/stdtypes.html#typesmapping) that maps [hashable](https://docs.python.org/3.10/glossary.html#term-hashable) values to arbitrary objects.  Dictionaries are sometimes found in other languages as "associative memories", "associative arrays" or "mappings".  Dictionaries differ from sequences, which are indexed by a range of numbers, in that dictionaries are indexed by *keys*, which can be any *hashable type (e.g. strings and numbers can always be keys). 

From the last paragraph of the definition of [hashable](https://docs.python.org/3.10/glossary.html#term-hashable).
```
Most of Pythons immutable built-in objects are hashable; mutable containers (such as lists or dictionaries) are not; immutable containers (such as tuples and frozensets) are only hashable if their elements are hashable. Objects which are instances of user-defined classes are hashable by default. They all compare unequal (except with themselves), and their hash value is derived from their id().
```

Also, check out [`__hash__()`](https://docs.python.org/3/reference/datamodel.html#object.__hash__) for more details.

Often dictionaries are thought of as a set of key:value pairs, with the requirement that keys are unique within a single dictionary.


## Creation

Dictionaries are defined initially using curly braces with zero or more key:value pairs separated by commas, or using  [dict()](https://docs.python.org/3.10/library/stdtypes.html#dict).   

````python
user_id = {'tom' : 12345, 'bob' : 67890}  # Create a dictionary using {} notation
user_id
````

````python
len(user_id)  # how many key:value pairs are in the dictionary
````

````python
dict([('tom', 12345), ('bob', 67890)])
````

When the keys are simple strings, a dictionary can be created using keyword arguments:

````python
dict(tom=12345, bob=67890)
````

## Indexing

Dictionaries are unordered, so they are indexed by *keys* and not indices.  However, the mechanism to index them is similar to other structures.  To access the values of a dictionary use square brackets with keys as "indices".

````python
user_id['bob']
````

````python
user_id['sam'] = 93846  # Add the 'sam':93846 key:value pair to the dictionary
user_id['sam']          # Get the value associated with the key 'sam'
````

````python
user_id['john']         # Accessing a key that does not exist throws an error
````

## Iteration

There are a few functions that can be used to iterate the elements of a dictionary.

````python
for k in user_id.keys():  # access all keys as an iterable
    print(k)
````

````python
for v in user_id.values():  # access all values as an iterable
    print(v)
````

````python
for k,v in user_id.items():  # access all key:value pairs as an iterable
    print(k, ':', v)
````

## The del Statement

Deleting key:value pairs from a dictionary can be done using the `del` statement:

````python
user_id
````

````python
del user_id['sam']
user_id
````

## Membership Testing

Membership tests can be perfomred using the `in` and `not in` keywords:

````python
user_id
````

````python
'tom' in user_id
````

````python
'john' in user_id
````

````python
12345 in user_id
````

````python
12345 in user_id.values()
````

# Dictionary Comprehensions

Dictionaries can also be created using comprehensions.  This is similar to lists and sets, and the notation more closely matches sets, however to differentiate a dictionary comprehension from a set comprehension the colon is used to declare the `key : value` pair in the target expression.

````python
{x : x**2 for x in (2,4,6)}
````

````python
names = ['tom', 'bob', 'sam']
ids = [12345, 67890, 93846]
{x : y for x,y in zip(names, ids)}
````
