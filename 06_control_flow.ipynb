# Indentation

Most programming languages have a mechanism for defining code blocks (grouping statements).  For example, C++ and other languages enclose blocks of code with `{` `}`.  Python does not do this and instead depends on indentation.  In Python we must tab (or use a consistent number of spaces) to indent and tell Python that we are defining a block.  The block begins with the first increased indentation and continue until the indentation decreases.  Generally, 4 spaces is preferred over tabs as tabstops may be modified by the editor and can make code more difficult to read/understand.

For example:

```
def func():
    statement 1
    statement 2

    statement 3
    
statement 4
```

In the above example, statements 1-3 are part of the function body and form a code block.  Statement 4 is part of the enclosing code block, and would not be executed as part of a call to `func()`. Functions will be covered in a later section.


# Control Flow

Control flow is the order in which statements (or instructions) are evaluated or executed.  Order of execution can be modified in several ways in Python, like many other languages.  In this section we will consider branching (if statements) and iteration (while and for loops).  



## `if` Statements

Similar to other languages Python supports the [`if` statement](https://docs.python.org/3/reference/compound_stmts.html#if) for branching.  

The `if` statement evaluates an expression and executes its body only if the expression evaluates to `True`.  The `else` is optional and its body is only executed if the corresponding `if` expression evaluates to `False`.  Additionally, one may use `elif`, which is a concatenation of `else` and `if` to evaluate additional expressions. `elif` is useful to avoid excessive indentation.

The `if` statement consists of the keyword `if` followed by an [expression](https://docs.python.org/3/reference/expressions.html) (some code that will be evaluated as `True` or `False`) and then by a colon `:`.  The body of the `if` begins on the next line and is indented.  `elif` has the same syntax as `if`, while `else` omits the expression.



````python
if 5 < 8:
    print('true')
````

````python
x = 0
if x < 2:
    print(x)
    x += 5
print(x)
````

The following example takes input from the user via `input`, which is returned as a string.  The input string is converted to an integer using `int` and stored in the variable `x`.  Note that not all string can be converted to integers, and so this operation may fail.  Feel free to test this here to see what happens.   If the conversion to integer succeeds, then the code will print different statements depending on the value of `x`.

````python
x = int(input("Please enter an integer: "))
if x < 0:
    x = 0
    print('Negative changed to zero')
else:
    if x == 0:
        print('Zero')
    else:
        if x == 1:
            print('One')
        else:
            print('More than one')
````

The same example as above is repeated below with `elif`. Notice how the indentation can get a bit carried away above, and is made more readable using `elif`.

````python
x = int(input('Please enter an integer: '))
if x < 0:
    x = 0
    print('Negative changed to zero')
elif x == 0:
    print('Zero')
elif x == 1:
    print('One')
else:
    print('More than one')
````

If you are familiar with the `switch` or `case` statements found in other languages, then the `if`...`elif`...`elif`... sequence serves as a substitute.  Alternatively, dictionaries can be used to emmulate this using the `get()` function, and in Python 3.10 [PEP 634](https://www.python.org/dev/peps/pep-0634/) now provides structural pattern matching which gives yet another means of emulating a switch statement.

````python
if x > 0: print('Positive')
````

## `while` Statements

loop_while.png

The [`while`](https://docs.python.org/3.8/reference/compound_stmts.html#while)  statement is used for repeated execution as long as an expression is `True`.  Compare this with the `if` statement which executes its body exactly once if its expression evaluates to `True`.

The `while` statement consists of the `while` keyword followed by an expression and a colon, similar to `if`.  The body should also appear indented on the following lines.

The following example outputs an initial sub-sequence of the [Fibonacci sequence](https://en.wikipedia.org/wiki/Fibonacci_number) using a while loop.

````python
# Fibonacci sequence:
# the sum of two elements defines the next
a, b = 0, 1
while a < 10:
    print(a, end=', ')
    a, b = b, a+b
````

The above example introduces a couple of new features:
- The first line contains a multiple assignment.  The variables `a` and `b` are simultaneously assigned the values `0` and `1`, respectively.  The same feature is used inside the loop body when assigning the new values of `a` and `b`.  It is important to note that in multiple assignments ALL of the expressions on the right-hand side are evaluated first (from left to right), BEFORE any of the assignments take place. 
- The `while` loop executes its body as long as the condition `a < 10` is `True`.  Recall the standard comparison operators are `<` (less than), `>` (greater than), `==` (equal to), `<=` (less than or equal to), `>=` (greater than or equal to) and `!=` (not equal to).    
- The `print` command includes an additional argument `end` to specify that ', ' follows the output instead of the default newline.  This results in the loop printing the Fibonacci series on the same line.

The reference page for the [`while`](https://docs.python.org/3.8/reference/compound_stmts.html#the-while-statement) statement gives complete details.

## `for` Statements

loop_for.png


The [`for`](https://docs.python.org/3.8/reference/compound_stmts.html#for) statement is slightly different from what you may have seen in C or other languges.  The Python `for` loop iterates over the items of any sequence (e.g. list, string) or other [iterable](https://docs.python.org/3/glossary.html?highlight=iterable#term-iterable) rather than requiring the definition of an iteration step and halting condition (as in C).  If you are unfamiliar with for loops in other languages do not worry, all you need to know is that this loop structure will execute the loop body once for each element of a given sequence, and when the sequence is exhausted the loop body will stop executing.

The `for` statement begins with the keyword `for` followed a variable, followed by the keyword `in`, followed by a sequence, and finally a colon `:`.  The variable in the `for` statement holds the current value from the sequence as the loop executes.

The following example will print each word and its length for the sequence `words`.

````python
words = ['plain', 'meadow', 'field']
for w in words:
    print(w, len(w))
````

The following will print each letter from the given string.

````python
# Iterate through a string
for letter in 'mountain':
    print(letter)
````

## `break`, `continue`, and `else` Statements

break_continue_else.png

The descriptions given above of `while` and `for` loops show only how to iterate until a condition fails or a sequence is exhausted, but what if we want to exit a loop early or perhaps skip particular iterations?  The statements discussed in this section will allow us to do exactly that.

### `break` Statments

Similar to C, `break` statements exit the innermost enclosing [for](https://docs.python.org/3/reference/compound_stmts.html#for) or [while](https://docs.python.org/3/reference/compound_stmts.html#while) loop.  

The following example consists of two nested `for` loops, both running from `0` to `3-1`.  The inner loop tests if its current value is greater or equal to the value of the outer loop, and if so calls the break statement exiting the inner loop.  Compare the output of this example with the next where the `break` statement is replaced by a `pass` statement (do nothing - See [pass Statements](#pass-Statements) below).

````python
for i in [0,1,2]:
    print('i =', i)
    for j in [0,1,2]:
        if j >= i:
            break
        print('  j =', j)
````

````python
for i in [0,1,2]:
    print('i =', i)
    for j in [0,1,2]:
        if j >= i:
            pass
        print('  j =', j)
````

### `continue` Statments

Also, similar to C, the `continue` statement skips to the next iteration of the innermost enclosing `for` or `while` loop.  

This example runs from `2` to `10-1` and outputs whether or not the number is even or odd.  Note that instead of using an `else` statement we are using `continue` to skip the remainder of the current iteration.  Compare the output of this example with the next where `continue` is omitted.

````python
for num in [2,3,4,5,6,7,8,9]:
    if num % 2 == 0:
        print(num, 'Even')
        continue
    print(num, 'Odd')
````

````python
for num in [2,3,4,5,6,7,8,9]:
    if num % 2 == 0:
        print(num, 'Even')
    print(num, 'Odd')
````

### `else` Clauses on Loops

Loop statements may have an `else` clause, which is executed when the loop terminates through the exhaustion of the iterable (with `for`) or when the condition becomes false (with `while`), but not when the loop terminates due to a `break` statement.

````python
for n in [2,3,4,5,6,7,8,9]:
    x = 2
    while x < n:   # A better way to implement this using `range()` appears later in the notebook
        if n % x == 0:
            print(n, '=', x, '*', n//x)
            x = x + 1
            break
        x = x + 1
    else:
        print(n, 'is a prime number')
````

Aside from there being a better implementation using `range()` (see below), the algorithm above is a brute force method that should only be used for very small numbers.  If you are interested in faster methods for finding primes, then check out the Sieve of [Eratosthenes, Sundaram, Atkin] and other methods.

## `pass` Statements

The `pass` statement does nothing.  It is used when a statement is required syntactically, but the program requires no action.

`while True:
    pass  # Busy-wait for keyboard interrupt (Ctrl+C)`
    
This loop will cycle forever unless it is interrupted.  `Ctrl+C` sends an interrupt that will cause the current task to abort, in this case stopping the while iteration and exiting the code.

`pass` is commonly used for creating minimal classes, or as placeholders for unfinished code.  

````python
class MyEmptyClass:
    pass

def initlog(*args):
    pass   # Remember to implement this!
````

## Looping Techniques

### Using `range()`

If you require to iterate over a squence of numbers you can use the built-in function [`range()`](https://docs.python.org/3.8/library/stdtypes.html#range).

````python
for i in [0,1,2,3,4,5]:
    print(i, '^2 = ', i**2, sep='')
````

````python
for i in range(6):
    print(i, '^2 = ', i**2, sep='')
````

Given a single argument `range()` will produce the sequence of integers from 0 to one less than the value of the argument.  Alternatively, you can specify the start value and a stop value.  Note that the sequence stops one before the given stop value.

````python
for i in range(4, 10):
    print(i, end=' ')
````

A step value may also be passed.  The returned values in the sequence are defined as `r[i] = start + step*i`, where `i >= 0` and `r[i] < stop`.

````python
for i in range(0,10,3):
     print(i, end=' ')
````

````python
for i in range(-50, -100, -20):
     print(i, end=' ')
````

If we revisit our prime example from above, a better way to implement this is using `range`.

````python
for n in range(2,10):
    for x in range(2,n):
        if n % x == 0:
            print(n, '=', x, '*', n//x)
            break
    else:
        print(n, 'is a prime number')
````

The following is copied from above just for comparison.  The nested for loops above using range perform the same operation as the loops (for/while) below.  However, range has the benefit of being easier to read/understand, and it allows us to more easily modify the values considered in the loops since we can use variables to define the range (as compared with the hard-coded lists of values that need to be updated manually).

````python
for n in [2,3,4,5,6,7,8,9]:
    x = 2
    while x < n:   # A better way to implement this using `range()` appears later in the notebook
        if n % x == 0:
            print(n, '=', x, '*', n//x)
            x = x + 1
            break
        x = x + 1
    else:
        print(n, 'is a prime number')
````

In the above examples `range()` behaves or appears to be a list, but if you attempt to print a range you will see something different:

````python
print(range(5))
````

The object returned by range is not a list.  It is an object that returns successive items of the desired sequence as you iterate over it without storing the actual list thus saving space.  The object is [iterable](https://docs.python.org/3.8/glossary.html#term-iterable), meaning that it is suitable as a target for functions and constructs that expect to obtain successive items until the supply is exhausted. 

````python
sum(range(4)) # 0 + 1 + 2 + 3
````

If you actually need or want a list from a range you can do the following:

````python
list(range(4))
````

You can also use the `len()` command with `range()` to iterate over the indices of a sequence.  

````python
a = ['This', 'is', 'not', 'the', 'way']
for i in range(len(a)):
    print(i, a[i])
````

### enumerate

To iterate through a sequence where both the index and value are needed use the [`enumerate()`](https://docs.python.org/3.8/library/functions.html#enumerate) function.

````python
a = ['This', 'is', 'the', 'way']
for i,v in enumerate(a):
    print(i, v)
````

### zip

To iterate through two or more sequences at the same time use the [`zip()`](https://docs.python.org/3.8/library/functions.html#zip) function.  zip will exit when at least one of the sequences is exhausted.

````python
questions = ['name', 'quest', 'favorite color']
answers = ['lancelot', 'the holy grail', 'blue']

for q,a in zip(questions, answers):
    print('What is your {0}?  It is {1}.'.format(q,a))  # similar to f'What is your {q}?  It is {a}.'
````

### reversed

To iterate through a sequence in reverse use the [`reversed()`](https://docs.python.org/3.8/library/functions.html#reversed) function.  This function does not modify the input sequence.

````python
for i in reversed(range(1,10,2)):
    print(i)
````

````python
basket = ['apple', 'orange', 'pear', 'banana']

for i in reversed(basket):
    print(i)
````

### sorted

To iterate through a sequence in sorted order use the [`sorted()`](https://docs.python.org/3.8/library/functions.html#sorted) function.  This function does not modify the input sequence.

````python
basket = ['apple', 'orange', 'pear', 'banana']

for f in sorted(basket):
    print(f)
    
print(basket)
````

### map

The `map()` function returns a map object (iterable) for the results after applyting the given function to each item of the given iterable (list, tuple, etc).

````python
def square(x):
    return x**2

for x in map(square, range(10)):
    print(x)
````
