## Built-in Help System

Similar to the `man` command in Linux/Mac, Python has a built-in function `help()` that provides implementation/use details.  For example to get more information on the `print` function enter:


```python
help(print)
```

If you are using Jupyter or the IPython shell you can also call up documentation using `?` after a function or variable.  Note that in Jupyter this may open a popup or separate window.

```python
print??
```

If you use `??` you can pull up the source code for many functions (built-in functions may only open the documentation).

## Documentation Strings ([Docstring](https://www.python.org/dev/peps/pep-0257/))

You may also see string literals wrapped with triple quotes (`'''` or `"""`) as the first statement (these may be multi-line) in a module, function, class or method definition.  These string literals are actually used to generate the documentation for the module, function, etc.  To view the *docstring* of a module, function, etc, type `help(<module name>)` at the prompt.  When you are finished reading the *docstring* press `q` to return back to the prompt.

Additionally, if you include docstrings in your own code, they will be used by the built-in help system.  In the following example, we create a very simple function that simply prints the argument passed to it.  The first lines of the function contain a docstring, that we will output using the `help` function.

```python
def ex_func(a):
    """
    This example function does nothing of particular use (other than print the
    argument passed to it), however it is being used to demonstrate the use of
    docstrings and the help system.
    
    a - A single value to be printed.  This variable is passed directly to the
        print function.
    """
    print(a)
    
help(ex_func)
```

## Comments

Many examples and professional programs contain non-executable code, also known as comments, that should give those reading the code some additional information that aids in the understanding of what the code is actually doing.  Comments in Python start with the hash character `#` and continue until the end of the line.  

```python
# this is a comment
a = 1  # this is another comment
print('a = ', a)
      # please make sure your comments are useful and meaningful
b = "# this is not a comment"
print(b)
```
